## Arithmetic in Python

Python can be used as a calculator. Try each of these:

```python
30 + 12        # Addition
```

```python
43 - 1         # Subtraction
```

```python
6 * 7          # Multiplication
```

```python
84 / 2         # Division — always returns a float!
```

```python
85 // 2        # Integer division
```

```python
7 ** 2         # Exponentiation
```

### Order of operations

Same as math (PEMDAS):

```python
(12 + 5) * 6   # Parentheses first → 102
```

```python
12 + 5 * 6     # Multiplication before addition → 42
```

## Data Types

Every value in Python has a **type**:

```python
type(42)       # int — integer (whole numbers)
```

```python
type(42.0)     # float — floating point (decimals)
```

```python
type('hello')  # str — string (text)
```

### Type conversion

You can convert between types:

```python
int(42.9)      # Truncates, does not round!
```

```python
float(42)
```

```python
str(42)
```

**Watch out:** division always returns a float:

```python
84 / 2         # Returns 42.0, not 42
```

## Strings

Text is enclosed in quotes (single `'...'` or double `"..."`):

```python
'Hello' + ' ' + 'World'   # Concatenation
```

```python
'Spam, ' * 3               # Repetition
```

```python
len('Portland')            # Length
```

`+` and `*` behave differently for strings vs. numbers:

```python
3 + 4          # Arithmetic → 7
```

```python
'3' + '4'      # Concatenation → '34'
```

**Common error** — you can't mix strings and numbers with `+`:

```python
'3' + 4        # TypeError! Try it and read the error message.
```
